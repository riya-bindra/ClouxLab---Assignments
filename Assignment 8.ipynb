### Problem: For a given set of values of x and y come up with your own way to predict the value of X 

Let us use Linear Regression to predict the values of X.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Data for training our model
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
```

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(x, y)

# Predict the value of X for a new value of x
new_x = np.array([[6]])
predicted_y = model.predict(new_x)
```

```python
print(f"The predicted value of y for x = 6 is: {predicted_y[0]}")
```

```python
# Plot the data and the regression line
plt.scatter(x, y, color='blue')
plt.plot(x, model.predict(x), color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
